# importing libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# getting data

```python
source=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source, 'lxml')
```

# getiing table from page cone and parsing it to list
Also we do same cleaning there - we got rid from rows with 'not assigned' borrows and added neighbourhoods same as brough to those rowa which had none

```python
table = soup.find("table")
res=[]
for row in table.find_all("tr"):
    cells = row.find_all("td")
    cells = [elem.text.strip() for elem in cells]
    if len(cells)==3:
        if cells[1]!='Not assigned':
            if cells[2]=='Not assigned':
                cells[2]=cells[1]
            res.append(cells)
            print(cells)
```

# gathering neighbourhoods with similar postcode

```python
lopi=[]
for i in range (len(res)):
    flag=0
    for j in range(len(lopi)):
        if (lopi[j][0]==res[i][0]) and (lopi[j][1]==res[i][1]):
            lopi[j][2]+=', '+res[i][2]
            flag=1
            break
    if flag==0:
        lopi.append(res[i])
    
```

# creating dataframe from list and saving it to disk

```python
df=pd.DataFrame(lopi)
df.columns=['PostalCode', 'Borough', 'Neighborhood']
df.to_csv('canada_postal_codes_m.csv')
df.head()
```
